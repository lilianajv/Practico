```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.lahaus.mx/venta/departamentos/quintana-roo"
req = requests.get(url)
soup=BeautifulSoup(req.text,"lxml")
```

```python
if None:
  print("not None")
else:
  print("esto es None")
```

```python
ids = []
bedrooms = []
bathrooms = []
parking_slots = []
area = []
precio = []

for article in soup.find_all("article", attrs={"class":"property-card-container"}):
  if article.find("ul"):
    
    for i, li in enumerate(article.find_all("li")):
       unit_details = li.find("div", attrs={"class": "flex justify-between pointer-events-none"})
       if unit_details is None:
        print("continue iteration")
        continue
       
       try:
        ids.append(article.get("real_estate_id") + "-" + str(i+1))
       except:
        ids.append(None)
       
       areas = li.find("p", attrs={"class": "text-14 text-carbon-800 text-carbon-light-800"}).text.strip()
       if not areas:
         areas = article.find("p", attrs={"label-r-12 text-carbon-light-800 mb-lh-8 pointer-events-none"})
       area.append(areas)

       price = li.find("p", attrs={"class": "small-r-14 text-carbon-800 mt-4 pointer-events-none"}).text.strip()
       if not price:
         price=article.find("p", attrs={"text-18 font-semibold text-carbon-800 mb-lh-8"})
       precio.append(price)

       hab_i = li.find("i", attrs={"class":"fa fa-real-estate-bedroom"}).find_parent("span").text
       #hab = hab_i.find("span", attrs={"class": "sr-only"})
       bedrooms.append(int(hab_i.split()[0][:1]))

       bano_i = li.find("i", attrs={"class": "fa fa-real-estate-bathroom"}).find_parent("span").text
       bathrooms.append(int(bano_i.split()[0][:1]))

       #park_i = li.find("i", attrs={"class": "fa fa-real-estate-parking"}).find_parent("span").text
       #if not park_i:
       #  park_i = None
       #parking_slots.append(park_i)

       



  else:
    ids.append(article.get("real_estate_id"))
    #precio
    price_2 = article.find("p", attrs={"class" : "text-18 font-medium mt-lh-4"}).text.strip()
    if not price_2:
      price_2 = article.find("p", attrs={"text-18 font-semibold text-carbon-800 mb-lh-8"})
    precio.append(price_2)
    #area
    areas = article.find("p", attrs={"text-14 leading-24"}).text.strip()
    if not areas:
      areas = article.find("p", attrs={"label-r-12 text-carbon-light-800 mb-lh-8 pointer-events-none"})
    area.append(areas.split()[:4])
    
    hab= article.find("p", attrs={"class" : "text-14 leading-24"}).find_parent("span")
    
    if not hab:
      hab=0
    bedrooms.append(hab)

    bano= article.find("p", attrs={"class" : "text-14 leading-24"}).find_parent("span")
    if not bano:
      bano = 0
    bathrooms.append(bano)

    ub=article.find("p", attrs={"class": "relative text-white pointer-events-none"}).text.strip()
    ubicacion.append(ub)
```
